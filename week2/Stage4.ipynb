```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plt.plot([x값], [y값]) -> 선그래프 그리기
# (1,90), (2,100), (3,50) 인 그래프 -> x축 값과 y축 값 따로 각각 모아서 적음
plt.plot([1,2,3],[90,100,50])
plt.show()
```

```python
plt.plot(['Math',' Korean', 'English'], [90, 100, 50])
plt.plot(['Math',' Korean', 'English'], [70, 60, 90])
plt.show()
# 두 개 그래프를 같이 그리면 구분되어져서 나옴
```

```python
plt.plot(['Math',' Korean', 'English'], [90, 100, 50])
plt.plot(['Math',' Korean', 'English'], [70, 60, 90])
plt.legend(['Choi','Kim'])

plt.show()
# plt.legend(['넣고 싶은거',넣고 싶은거']) -> 옆에 박스가 표시되어 구분 됨
```

# 판다스로 csv 데이터 불러오기

```python
import pandas as pd
train = pd.read_csv('data/train.csv')

train.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# sns.relplot(data = 데이터명, x = 'x축 데이터', y = 'y축 데이터') -> 산포도

sns.relplot(data=train, x = 'Age', y ='Survived')
```

```python
#  aspect = n -> 가로 길이가 세로길이의 n배
#  hue ='원하는 기준' => 해당 기준에 따라 색이 다르게 표시됨.
# ex) hue = 'Sex' -> 성별에 따라 색이 다르게 표시됨.
sns.relplot(data=train, x = 'Age', y ='Survived', hue= 'Sex', aspect = 2 )
```

```python
# sns.catplot(data=데이터명, x='x축 데이터', y= 'y축 데이터') -> catplot 그리기
sns.catplot(data=train, x='Pclass', y= 'Fare', aspect = 2, hue ='Sex' )
```

```python
# sns.catplot(data=데이터명, x='x축 데이터', y= 'y축 데이터') -> catplot 그리기
sns.catplot(data=train, x='Pclass', y= 'Fare', aspect = 2, jitter = False )
# jitter을 False로 설정하지 않는 이상 기본 값인 True로 설정되어 있음
```

```python
sns.pairplot(data=train, hue = 'Sex',    # hue = 'Sex' -> 성별을 기준으로 색표시
            x_vars=['Pclass', 'Age', 'Fare'],
            y_vars=['Survived','Age', 'Fare'], height=3 )

# pairplt(data= 데이터파일명, 
#          x_vars=['x축으로 삼고 싶은 컬럼'],
#          y_vars=['y축으로 삼고 싶은 컬럼'])
```

```python
# boxplot그리기 
# sns.catplot(data = 데이터명, x= 'x축', y= 'y축', kind=)
sns.catplot(data=train, x='Pclass', y= 'Fare', hue='Sex', kind='box')

# 그래프 밖의 점은 유효하지 않은 값
# 선의 끝 -> 각 최소 최대
# 가운데 선 -> 중앙값
# 박스의 아래 -> 1분위수(오름차순 정렬시 25%)
# 박스의 위 -> 3분위수(오름차순 정렬시 75%) 
```
