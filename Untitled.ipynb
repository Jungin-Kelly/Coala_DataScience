```python
import pandas as pd
df = pd.read_csv('2015.csv')
df
movietitle = df['영화제목']
print(movietitle)
    
```

```python
# 네이버 영화 데이터 수집
import requests
from bs4 import BeautifulSoup

for name in movietitle:
    url = "https://movie.naver.com/movie/search/result.nhn?query="+name+"&section=all&ie=utf8"
    
    raw = requests.get(url ,headers ={"User-Agent":"Mozilla/5.0"})

    html = BeautifulSoup(raw.text, 'html.parser')

    #컨테이너 ul.search_list_1
    movies = html.select_one("ul.search_list_1 dl")
    title = movies.select_one("dl>dt").text  #'.text' 해야지 테그가 안딸려옴.
    score = movies.select_one("em.num")
    print(name)
    if not score:
        print("none")
    else : print ("평점:",score.text)
    
    #장르
    genre = movies.select_one("ul.search_list_1 dl dd.etc:nth-of-type(2)") #뒤에 a까지 쓸 수 있음 -> 첫번째 a라는 뜻
    all_genre = genre.text
    genre_list = all_genre.split('|')
    print(genre_list)
    
    #감독, 배우
    director = movies.select_one("ul.search_list_1 dl dd.etc:nth-of-type(3)")
    if director is None:
        all_director = 'none...'
        print("none")
    else:
        all_director = director.text
        director_list = all_director.split('|')
        print(director_list)
      
  

        
  
#     for g in genre:                        # Q. 출력시 배우: <- 얘는 한 번만 나오게 하는 법?
#         print("장르:", g.text)
#     for d in director:
#         print("감독, 배우:", d.text)
    print("=" * 50)
    
#     print(,score)
```

```python
# 크롤링을 위한 기본 세팅

import pandas as pd
df = pd.read_csv('data_final.csv')
df
movietitle = df['name']
import openpyxl
wb=openpyxl.Workbook()
ws = wb.active

import requests
from bs4 import BeautifulSoup


for name in movietitle:
    write=[]
    url = "https://movie.naver.com/movie/search/result.nhn?query="+name+"&section=all&ie=utf8"
    
    raw = requests.get(url ,headers ={"User-Agent":"Mozilla/5.0"})

    html = BeautifulSoup(raw.text, 'html.parser')

    

    
    #컨테이너 : ul.search_list_1 dl
    
    movies = html.select_one("ul.search_list_1 dl")
    
    #제목 
    name = movies.select_one("dl>dt").text  
    write.append(name)
    
    #평점
    score = movies.select_one("em.num")
    
    if not score:
        write.append("None")
    else : 
        write.append(score.text)
    
    
    
    #장르
    genre = movies.select_one("ul.search_list_1 dl dd.etc:nth-of-type(2)")
    all_genre = genre.text
    genre_list = all_genre.split('|')
    print(genre_list)
    write.extend(genre_list)
    
    #감독&배우
    director = movies.select_one("ul.search_list_1 dl dd.etc:nth-of-type(3)")
    if director is None:
        all_director = 'none...'
        print("none")
        write.append("None")
    else:
        all_director = director.text
        director_list = all_director.split('|')
        print(director_list)
        write.extend(director_list)
    ws.append(write)

# 파일 저장하기    
wb.save("평점감독배우장르.xlsx")
```
