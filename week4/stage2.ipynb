```python
import pandas as pd
```

```python
df = pd.read_csv('data/iris.csv')
df.head(10)
```

```python
df.info() # 해당 데이터 프레임의 자료형이 나옴.
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

```python
# catplot 으로 꽃받침의 길이와 너비로 종 분포도 보기
sns.catplot(data= df, x='SepalLengthCm', y = 'SepalWidthCm', hue = 'Species', aspect =2)
```

```python
# catplot으로 꽃잎의 길이와 너비로 종 분포도 보기 
sns.catplot(data= df, x='PetalLengthCm', y = 'PetalWidthCm', hue = 'Species', aspect =2)
```

```python
# 꽃받침의 길이에 따른 밀집도
facet = sns.FacetGrid(df,hue='Species', aspect= 2)
facet.map(sns.kdeplot, 'SepalLengthCm')
facet.add_legend()
plt.show()
```

```python
sns.catplot(kind='box', data = df, x = 'Species', y= 'SepalLengthCm', aspect =2)
```

```python
# violin plot - x:종, y: 꽃받침 길이
sns.violinplot(data=df, x= 'Species', y='SepalLengthCm')
```

```python
# violin plot - x:종, y: 꽃받침 너비
sns.violinplot(data=df, x= 'Species', y='SepalWidthCm')
```

```python
# violin plot - x:종, y: 꽃잎 길이
sns.violinplot(data=df, x= 'Species', y='PetalLengthCm')
```

```python
# violin plot - x:종, y: 꽃잎 너비
sns.violinplot(data=df, x= 'Species', y='PetalWidthCm')
```

# plot별 특징
- boxplot -> 최대 최소 등 값의 처음과 끝을 알 수 있음.
- kdeplot -> 분포의 밀집도를 알 수 있음
- violinplot ->  box plot과 kdeplot의 장점을 합쳐놓은 것/ 값의 처음과 끝과 밀집도 확인0
