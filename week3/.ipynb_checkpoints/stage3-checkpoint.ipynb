# Train set vs Test set
train set :(훈련용) 데이터의 특성을 살펴보고 숨겨진 규칙 및 패던을 찾기 위한 데이터  
test set : (시험용) 찾아낸 패턴이 정말 맞는지 확인하기 위한 시험용 데이터


```python
df['Age'].fillna(df['Age'].mean())
# Nan : 값이 없음
#fillna(넣고 싶은 값) -> Nan의 값을 원하는 값으로 대체해주는 함수
# ***fillna는 값을 영구적으로 바꿔주는게 아님. 그 순간에만 바꿔져 보이게 해줌
# 만약 값을 영구적으로 바꾸고 싶으면 값을 재정의 해야함. 
# df['Age'] = df['Age'].fillna(df['Age'].mean())  ->이렇게
```

```python
df['Age'].mean()
#'Age'의 평균값
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean()) 
```

```python
df
```

```python
df['Age'] 
#위에서 fillna한 값으로 age 값을 재정의 해줬기 때문에 NAN값이 바뀌었음을 확인
```

```python
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean()) 
# textset 에도 동일한 변형을 적용해야함. b/c 하나의 모델을 사용할 것이기 때문
```

```python
# df.loc[조건, 열] = 넣고 싶은 값
df.loc[ df['Age'] < 10, 'Age'] = 0
df.loc[ (df['Age'] >= 10) &  (df['Age'] < 20), 'Age' ] = 1
df.loc[ (df['Age'] >= 20) &  (df['Age'] < 30), 'Age' ] = 2
df.loc[ (df['Age'] >= 30) &  (df['Age'] < 40), 'Age' ] = 3
df.loc[ (df['Age'] >= 40) &  (df['Age'] < 50), 'Age' ] = 4
df.loc[ df['Age'] >= 50, 'Age'] = 5
```

```python
df
```

```python
# df_test.loc[조건, 열] = 넣고 싶은 값
df_test.loc[ df_test['Age'] < 10, 'Age'] = 0
df_test.loc[ (df_test['Age'] >= 10) &  (df_test['Age'] < 20), 'Age' ] = 1
df_test.loc[ (df_test['Age'] >= 20) &  (df_test['Age'] < 30), 'Age' ] = 2
df_test.loc[ (df_test['Age'] >= 30) &  (df_test['Age'] < 40), 'Age' ] = 3
df_test.loc[ (df_test['Age'] >= 40) &  (df_test['Age'] < 50), 'Age' ] = 4
df_test.loc[ df_test['Age'] >= 50, 'Age'] = 5
#edit 에서 find and replace 기능을 활용하면 빠르게 바꿀 수 있음.
```

```python
# Sibsp + Parch 로 FamilySize 만들기
df['FamilySize'] = df['SibSp'] + df['Parch']
df.head()
```

```python
# test set에도 동일하게 적용
df_test['FamilySize'] = df_test['SibSp'] + df_test['Parch']
df_test.head()
```

```python
# 특정 값으로만 새로운 데이터 프레임 만들기
train = df[['Survived', 'Sex', 'Age', 'FamilySize']]
test = df[['Sex', 'Age', 'FamilySize']]
train.head()
```

```python
# train과 test를 재정의 해야지 앞에서 변경한 값들이 반영됨.
train = df[['Survived', 'Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
test = df_test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
train.head()
```

# stage3

```python
# 여러 개 항목을 선택할 때는 대괄호 2개/ 한 개 항목만 선택할 때는 대괄호 1개
# x_train, y_train -> train에서 데이터를 가져왔기 때문
# skitlearn쓸 때는 아래와 같이 데이터를 2부분으로 나눠야함.
x_train = train[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
y_train = train['Survived']

x_train
```

```python
y_train
```

```python
# tab 누르면 자동완성됨.
# sklearn은 기능이 많기 때문에 디시젼트리에 해당하는 기능만 가져옴
# 1주차의 황금계수를 찾는 과정이라 보면됨.
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()  # 디시전트리를 초기화함
tree.fit(x_train, y_train)       # 학습시킴

tree.score(x_train, y_train)     # 모델의 정확도를 판단
# 우리의 트리가 train 데이터 기준으로 94퍼의 정확도를 가짐
```

```python
x_test = test[['Sex', 'Age', 'FamilySize', 'Fare', 'Embarked']]
x_test
```

```python
# sklearn의 예측기능 사용하기
prediction = tree.predict(x_test)
prediction
```

```python
# 위에 나온 값은  x_test에서 survived 값을 예측한 결과임
x_test 
```

```python
# 제출용 새로운 데이터 프레임을 만듦
sumit = pd.DataFrame({
     'PassengerId': df_test['PassengerId'],  #df_test에서 'PassengerId'만 떼와서 저장
    'Survived': prediction  # survived는 앞에의 prediction 값을 가져와서 저장
})

# sumit.to_csv('파일명.csv', index = False) -> csv파일로 저장하기
# kaggle은 인덱스 번호가 없는 파일을 원하기 때문에 index = False로 함.
# index 번호 붙이려면 index = True 하면 됨.
sumit.to_csv('sumit.csv', index = False)
```
